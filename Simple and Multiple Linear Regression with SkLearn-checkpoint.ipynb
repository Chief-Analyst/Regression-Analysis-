```python
##Mulptiple Linear Regression

##Import relevant libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
##Load Data

data = pd.read_csv("C:/Users/Administrator/Downloads/real_estate_price_size_year_view.csv")
data.head()
```

```python
##Creating dummy variables for 'view' 

data["view"] = data["view"].map({"No sea view": 0, "Sea view": 1})
```

```python
data.head()
```

```python
data.describe()
```

```python
##Declare dependent and independent varaibales

x = data[['size', 'year', 'view']]
y = data['price']
```

```python
##Regression

reg = LinearRegression()
reg.fit(x,y)
```

```python
##Coefficients

reg.coef_
```

```python
##Intercept

reg.intercept_
```

```python
##R-squared

reg.score(x,y)
```

##Formula for adjusted R-square

$R^2_{adj.} = 1 - (1-R^2)*\frac{n-1}{n-p-1}$

```python
x.shape
```

```python
##Adjusted R-square

r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
```

```python
p_values = f_regression(x,y)[1]
p_values
```

```python
p_values.round(3)
```

```python
##Creating a summary table

reg_summary = pd.DataFrame(data = x.columns.values, columns=['Features'])
reg_summary
```

```python
reg_summary ['Coefficients'] = reg.coef_
reg_summary ['p-values'] = p_values.round(3)
reg_summary
```

```python
##Checking to see how much significance 'year' has on the model.
##To determine if I am dropping it or not.

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
import numpy as np

model_all = LinearRegression().fit(x, y)
r2_all = np.mean(cross_val_score(model_all, x, y, cv=5, scoring='r2'))

x_no_year = x.drop(columns=['year'])
model_no_year = LinearRegression().fit(x_no_year, y)
r2_no_year = np.mean(cross_val_score(model_no_year, x_no_year, y, cv=5, scoring='r2'))

print("R² with year:", r2_all)
print("R² without year:", r2_no_year)
```

```python
##Standardization

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

```python
##Regression with scaled features

reg = LinearRegression()
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
##Creating a summary table

reg_summary = pd.DataFrame([['Intercept'], ['size'], ['year'], ['view']], columns = ['Features'])
reg_summary['Weights'] = [reg.intercept_, reg.coef_[0], reg.coef_[1], reg.coef_[2]]
```

```python
reg_summary
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

r2_scores = cross_val_score(reg, x_scaled, y, cv=5, scoring='r2')
print("Average R² across folds:", np.mean(r2_scores))
```

```python
reg.predict(new_data)
```
