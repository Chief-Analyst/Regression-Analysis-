```python
##FIRST MULTIPLE LINEAR REGRESSION 
##DATA IS A FICTICIOUS REAL ESTATE DATA 

##Importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
##Load the data into a dataframe

sea_view = pd.read_csv("C:/Users/Administrator/Downloads/real_estate_price_size_year_view.csv")
```

```python
sea_view.describe()
```

```python
print(sea_view.head())
```

```python
sea_view_data = sea_view.copy()
```

```python
##Creating dummy variables for 'view' 

sea_view_data["view"] = sea_view_data["view"].map({"No sea view": 0, "Sea view": 1})
```

```python
sea_view_data.head()
```

```python
##Creating multiple regression

y = sea_view_data['price']
x1 = sea_view_data[['size', 'year', 'view']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
##Plot the results

sea_view_data.describe()

```

```python
##Predicted values
y_hat_no = 74152.54 + 223.0316 * sea_view_data['size']
y_hat_yes = 130882.54 + 223.0316 * sea_view_data['size']

##Scatter plot of actual data
plt.scatter(sea_view_data['size'], sea_view_data['price'], color='blue', label='Actual Data')

##Plot the regression lines
plt.plot(sea_view_data['size'], y_hat_no, color='red', label='No Sea View')
plt.plot(sea_view_data['size'], y_hat_yes, color='green', label='Sea View')

##Labels and title
plt.xlabel('Size')
plt.ylabel('Price')
plt.title('Regression Lines for Sea View vs No Sea View')
plt.legend()
plt.show()

```

```python
##Making predictions based on the regression
##Trying to predict price with a size of 1500 with no sea view from year 2010 and a size of 1000 with sea view from same year.

new_data = pd.DataFrame({'const':1, 'size':[1500,1000], 'year':[2010,2010], 'view':[0,1]})
new_data = new_data[['const', 'size', 'year', 'view']]
new_data.rename(index={0:'no', 1:'yes'})

```

```python
predictions = results.predict(new_data)
predictions
```

```python
predictionsdf = pd.DataFrame({'Predictions':predictions})
joined = new_data.join(predictionsdf)
predictionsdf
```
